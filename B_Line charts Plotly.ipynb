```python
# import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd


trace_meth = go.Scatter(
                x=df['year'],
                y=df['meth'],
                name = "Methamphetamine",
                line = dict(color = '#8BBBD9',
                           width = 4),
                opacity = 0.8)

trace_heroin = go.Scatter(
                x=df['year'],
                y=df['heroin'],
                name = "Heroin",
                line = dict(color = '#2987D9',
                           width = 4),
                opacity = 0.8)

trace_opiates = go.Scatter(
                x=df['year'],
                y=df['marij'],
                name = "Marijuana",
                line = dict(color = '#A0D9A4',
                           width = 4),
                opacity = 0.8)

trace_cocaine = go.Scatter(
                x=df['year'],
                y=df['cocaine'],
                name = "Cocaine",
                line = dict(color = '#B5BDBD',
                           width = 4),
                opacity = 0.8)

trace_marij = go.Scatter(
                x=df['year'],
                y=df['opiates'],
                name = "Opiates",
                line = dict(color = '#012340',
                           width = 4),
                opacity = 0.8)

data = [trace_cocaine, trace_marij, trace_meth, trace_opiates, trace_heroin]

layout = dict(
    title = "Drug Popularity over Time as Indicated by Substance Abuse Treatment Admissions")

fig = dict(data=data, layout=layout)
pyo.iplot(fig)
# py.iplot(fig, filename = "Manually Set Range")
```
